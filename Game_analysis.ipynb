#### Восстановление года релиза игры, заполнение пропусков

```python
count=0
def find_year(row):
    global count
    string = row['name']
    
    for word in string.split():
        match = re.search(r'(\d{4})', word)# вытаскиваем комбинацию из 4-х цифр - год.
        if match:
            year = match.group()
            if int(year) > 2017:
                year = 0
            count +=1
        else:
            year = 0
    return year
# вытаскивать комбиации годов: 89,99 не стал, случаи совсем уж единичные.
```

```python
def insert_year(row):
    global count2                        
    year_rel=row['year_of_release']
    if ((type(year_rel) == float) and( pd.isna(year_rel))):
        year = row['year']
        if int(year)>0:
            count2 +=1
            return year
        else:
            year = row['year_of_release']
            return year
    else:
        year = row['year_of_release']
        return year
```

```python
count2=0
df['year']=df.apply(find_year , axis = 1)
print (f'Восстановлено {count} годов-релизов.')
```

```python
df['year']=df['year'].astype(int)
```

```python
df['year_of_release']=df.apply(insert_year,axis = 1)
#df.loc[df['year_of_release'].isna() , 'year_of_release'] = 0
print (f'Фактически восстановлено {count2} "год релиза".')
```

```python
# На тепловой кате посмотрим зависимости
# интересует корреляция между оценкой критикой и продажами, она есть (15-17%) это должно "пригодиться" в ML
cols = ['year_of_release', 'critic_score','na_sales','eu_sales','jp_sales']
heatmap = sns.heatmap(df[cols].corr(), cbar=True, annot=True,cmap='BrBG');
heatmap.set_title('Тепловая карта корреляции', fontdict={'fontsize':18}, pad=12);
plt.xticks(rotation=45, ha='right');
```

```python
# здесь создаем ДФ на котором будем обучаться, он без пропусков! 7000 значений, есть на чем попробовать!
df_train = df[((~df['critic_score'].isna()) & (~df['user_score'].isna()))]
df_train
```

```python
# здесь создаем ДФ на котором применим обученную модель 
df_test = df[((~df['critic_score'].isna()) & (df['user_score'].isna()))]
df_test
```

```python
# создаем обучающие массивы.
X_df_train = df_train[['year_of_release','critic_score', 'na_sales','eu_sales','jp_sales']].values 
Y_df_train = df_train['user_score'].values  

model_df = RandomForestClassifier(n_estimators=300).fit(X_df_train,Y_df_train)
# точность прогнозирования, чем больше к единице, тем лучше
r_sq_df = model_df.score(X_df_train,Y_df_train)
r_sq_df
```

```python
# функция выводит тепловую карту корреляции для конкретной платформы
def map_corr (pltform):
    cols = ['critic_score','user_score','total_sales']
    df_corr = df[((df['platform']==pltform) & (df['year_of_release']>2012) &  (df['critic_score']!=0))]
    heatmap = sns.heatmap(df_corr[cols].corr(), cbar=True, annot=True,cmap='BrBG');
    heatmap.set_title(f'Тепловая карта корреляции по консоле {pltform}', fontdict={'fontsize':18}, pad=12);
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0, ha='right');
```

```python
# функция выводит график зависимости объема продаж от оценки критиков/пользователей
# для конкретной платформы
def plot_corr_sct (pltform):
    pltf = df[(df['platform'] == pltform)][['total_sales', 'critic_score', 'user_score']]
    ax = pltf.plot.scatter( x='critic_score', y='total_sales',figsize=(15,8), alpha=0.9, color='red')
    plt_sct = pltf.plot.scatter(
        x='user_score', y='total_sales', figsize=(15,8), grid=True,alpha=0.5,
        title=f'Зависимость объемов продаж игр от оценки критиков/пользователей для консоли {pltform}',
        color='darkblue',ax=ax)
    plt_sct.set_xlabel('Оценка критиков/пользователей',fontsize=14)
    plt_sct.set_ylabel('Объемы продаж, млн.шт',fontsize=14)
    plt_sct.legend(['Оценка критиков','Оценка пользователей'], fontsize=14);
```

```python
# функция формирует гситограмму по Оценкам критикам/пользователей, для конкретной платформы.
def count_score(pltform):
    pltf2 = df[(df['platform'] == pltform)][['total_sales', 'critic_score', 'user_score']]
    ax = pltf2.plot(y = 'critic_score', kind = 'hist', bins = 25, grid=True, figsize = (10,3), range = (40,100))
    pltf2.plot(y = 'user_score', kind = 'hist', bins = 25, grid=True, figsize = (10,3), range = (40,100),ax=ax, alpha = 0.4)
    plt.title('Гистограмма Оценки критиков/пользователей. ')
    plt.xlabel('Оценка критиков/пользователей ')
    plt.ylabel('Кол-во игр с подобной оценкой')
    plt.legend(['Оценка критиков','Оценка пользователей'], fontsize=12, loc='upper left');
    plt.show()
```

##  Формирование портрета пользователя для каждого региона (NA, EU, JP)

**TOP 5 платформ определим в рамках ранее утвержденного временного интервала - 5 лет.**

```python
# в данной функции формируем и выводим на круговую диаграмму ТОП 5 популярныъ жанров и платформ (для сравнения)
def pie_region(region):
    region_platform_pivot = (df[df['year_of_release'] > 2012]
                         .pivot_table(index = 'platform', values =['na_sales', 'eu_sales', 'jp_sales'],aggfunc = 'sum'))
    region_genre_pivot = (df[df['year_of_release'] > 2012]
                         .pivot_table(index = 'genre', values =['na_sales', 'eu_sales', 'jp_sales'],aggfunc = 'sum'))
    region_platform_top =region_platform_pivot.loc[:,region].sort_values(ascending=False).head(5).reset_index()
    region_genre_top =region_genre_pivot.loc[:,region].sort_values(ascending=False).head(5).reset_index()

    region_platform_top['%'] = region_platform_top[region] / region_platform_top[region].sum()
    region_genre_top['%'] = region_genre_top[region] / region_genre_top[region].sum()
    
    if region == 'eu_sales':
        rname = 'Европейском'
    elif region == 'na_sales':
        rname = 'Северо-Американском'
    else:
        rname = 'Японском'
        
    explode = (0.1, 0.1,0.1,0,0)

    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(15, 5))
    ax1.pie(region_platform_top[region], shadow=True,
        labels=region_platform_top['platform'],explode=explode,
        startangle=130, autopct='%1.1f%%',textprops=dict(color='darkblue',weight='bold',fontsize=12))
    
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax1.set_title(f'Top5 Платформ на {rname} рынке',fontsize=12,weight='bold')

    ax2.pie(region_genre_top[region], shadow=True,
        labels=region_genre_top['genre'],explode=explode,
        startangle=130, autopct='%1.1f%%',textprops=dict(color='darkblue',weight='bold',fontsize=12))

    ax2.set_title(f'Top5 Жанров на {rname} рынке',fontsize=12,weight='bold');
```

```python
# в данной функции выводим продажи игр по возратным категориям 
def bar_esrb(region):
    df_2012=df[df['year_of_release'] > 2012]
    df_rating = df_2012.pivot_table(index='rating_full', values=region, aggfunc='sum')
    df_rating['%']=df_rating[region]/df_rating[region].sum()
    print (df_rating)
    my_colors = list(islice(cycle(['r', 'g', 'b', 'c', 'm', 'y']), len(df))) 
    df_rating[region].plot(kind = 'bar',figsize = (7,3),grid=True, color = my_colors )
    plt.title('Возрастная категория игр и их объемы продаж')
    plt.xlabel('Возрастной рейтинг компьютерных игр')
    plt.ylabel('Объемы продаж\n для данной возрастной категории игр')
    plt.xticks(rotation=45); 
```

H₀ - Средние пользовательские рейтинги платформы Xbox One  == Средние пользовательские рейтинги платформы PC
H₁ - Средние пользовательские рейтинги платформы Xbox One  !== Средние пользовательские рейтинги платформы PC

Для проверки вышеозвученных допущений, воспользуемся методом для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind с параметром equal_var раным False т.к. чуть ниже мы видим, что размер совокупностей **сильно** отличается и дисперсии у совокупностей пусть не сильно но разняться.


```python
ax = plt.gca()
xone = df.query('platform == "xone"')['user_score']
pc = df.query('platform == "pc"')['user_score']

xone.plot(kind='hist',ax=ax, color='green',figsize=(10,5),range = (20,100))
pc.plot(kind='hist',ax=ax, color='lightblue',alpha = 0.7,figsize=(10,5),range = (20,100))

plt.axvline(x=xone.median(),linewidth=3, color='darkgreen', linestyle=':')
ax.text(xone.median()-7, 200, '{0:.1f}'.format(xone.median()),fontsize='large')

plt.axvline(x=pc.median(),linewidth=2, color='blue', linestyle='--')
ax.text(pc.median()+1, 200, '{0:.1f}'.format(pc.median()), fontsize='large')

print ('Размер совокупности user_score для консоли Xbox One:',len(xone))
print ('Размер совокупности user_score для консоли PC:',len(pc))
print ('Расчет Дисперсии для Xbox One =','{0:.2f}'.format(xone.var()))
print ('Расчет Дисперсии для PC =','{0:.2f}'.format(pc.var()))
print ('Разница между Дисперсиями:','{:.2f}%'.format(100-100*xone.var()/pc.var()))

plt.legend(['Медиана Xbox One ', 'Медиана PC', 'Xbox One', 'PC'], prop={'size':12});
plt.title('Гистограмма Оценки пользователей для игр на платформах Xbox One и PC.')
ax.set_xlabel('Оценка пользователей')
ax.set_ylabel('Частота');

```

```python
alpha = .01 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(xone,pc,equal_var=False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

### Гипотеза №2: Средние пользовательские рейтинги жанров Action  и Sports  разные.

Проверка гипотезы №2
• Средние пользовательские рейтинги жанров Action и Sports разные.  

Определим нулевую и альтернативную гипотезу.
H₀ - Средние пользовательские рейтинги жанров Action и Sports - РАВНЫ.
H₁ - Средние пользовательские рейтинги жанров Action и Sports - НЕ РАВНЫ.
Действуем по аналогии, применим специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind (array1, array2,equal_var), но здесь параметр equal_var будет True, т.е. не будем его использовать.  

(Дисперсии практически равны, размер совокупностей отличаются не в 4 раза как ранее)

```python
ax = plt.gca()
action = df.query('genre == "action"')['user_score']
sports = df.query('genre == "sports"')['user_score']

action.plot(kind='hist',ax=ax, color='green',figsize=(10,5),range = (20,100))
sports.plot(kind='hist',ax=ax, color='lightblue',alpha = 0.7,figsize=(10,5),range = (20,100))

plt.axvline(x=action.median(),linewidth=3, color='darkgreen', linestyle=':')
ax.text(action.median()+2, 300, '{0:.1f}'.format(action.median()),fontsize='large')

plt.axvline(x=sports.median(),linewidth=2, color='blue', linestyle='--')
ax.text(sports.median()-5, 200, '{0:.1f}'.format(sports.median()), fontsize='large')

print ('Размер совокупности user_score для жанра Action:',len(action))
print ('Размер совокупности user_score для жанра Sports:',len(sports))
print ('Расчет Дисперсии для жанра Action =','{0:.2f}'.format(action.var()))
print ('Расчет Дисперсии для жанра Sports =','{0:.2f}'.format(sports.var()))
print ('Разница между Дисперсиями:','{:.2f}%'.format(100-100*action.var()/sports.var()))

plt.legend(['Медиана жанра Action ', 'Медиана жанра Sports', 'Action', 'Sports'], prop={'size':12});
plt.title('Гистограмма Оценки пользователей для жанров Action и Sports.')
ax.set_xlabel('Оценка пользователей')
ax.set_ylabel('Частота');
```

```python
alpha = .01 # критический уровень статистической значимости

results = st.ttest_ind(action,sports)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

# Гипотеза о том, что Средние пользовательские рейтинги жанров Action и Sports равны - не подтвердилась!

```python
# Решил проверить, гипотезу равенства дисперсий для совокупностей Экшен и Спорт....
#H0 - Дисперсии не равны
#H1 - Дисперсии равны
d1=action.describe()[2]**2
d2=sports.describe()[2]**2
df1 = len(action)-1
df2 = len(sports)-1
alpha = 0.05

results = st.f.sf(d1/d2, df1, df2)

print('p-значение:', results)

if (results < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('equal_var = True, т.е. как параметр его не используем')
```
